```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Let us create our own parabolic function

```python
samples = 100
X = np.linspace(0, 10, 100)
y = X ** 3 + np.random.randn(samples) * 100 + 100
plt.figure(figsize=(8,7))
plt.scatter(X, y)
```

Checking whether Linear Regression model will fit to this kind of data.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr = LinearRegression()
lr.fit(X.reshape(-1, 1), y)
y_pred = lr.predict(X.reshape(-1,1))

plt.scatter(X, y);
plt.plot(X, y_pred);
print('R2 SCORE',r2_score(y, y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y, y_pred))) 
```

Now let us check with Polynomial Regerssion

```python
# Using degree = 2 for the parabolic function

from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures(degree=2)
X_poly = pr.fit_transform(X.reshape(-1, 1))
lr = LinearRegression()
lr.fit(X_poly, y.reshape(-1, 1))
y_pred = lr.predict(X_poly)
plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, y_pred);

print('R2 SCORE',r2_score(y, y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y, y_pred))) 
```

```python
# Using degree = 2 for the parabolic function

pr = PolynomialFeatures(degree=3)
X_poly = pr.fit_transform(X.reshape(-1, 1))
lr = LinearRegression()
lr.fit(X_poly, y.reshape(-1, 1))
y_pred = lr.predict(X_poly)
plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, y_pred);

print('R2 SCORE',r2_score(y, y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y, y_pred))) 
```

```python
# Using degree = 2 for the parabolic function

pr = PolynomialFeatures(degree=4)
X_poly = pr.fit_transform(X.reshape(-1, 1))
lr = LinearRegression()
lr.fit(X_poly, y.reshape(-1, 1))
y_pred = lr.predict(X_poly)
plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, y_pred);

print('R2 SCORE',r2_score(y, y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y, y_pred))) 
```

The r2 score improved a lot from linear regression's 0.7389 to polynomial regression's 0.9026.
You can choose either 3 or 4 as the polynomial degree for the function as it is giving almost the same.
